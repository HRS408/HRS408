```python
mask = vehicles["citympg"].isnull()
vehicles["citympg"] = vehicles["citympg"].fillna(vehicles["citympg"].mean())
```

```python
mask = vehicles["highwaympg"].isnull()
vehicles["highwaympg"] = vehicles["highwaympg"].fillna(vehicles["highwaympg"].mean())
```

```python
response = "co2emissions"
y = vehicles[response]
y
```

```python
iVariables = ['citympg','highwaympg','displacement']
x = vehicles[iVariables]
# mask = x.isna()
# x = x.fillna(x['citympg'].mean())
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1234)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```
